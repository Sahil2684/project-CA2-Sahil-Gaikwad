```python
import matplotlib.pyplot as plt
import seaborn as sns
CO_data = df['CO']

mean = CO_data.mean()
median = CO_data.median()
mode = CO_data.mode()[0]
std_dev = CO_data.std()

print(f"Mean CO: {mean}")
print(f"Median CO: {median}")
print(f"Mode CO: {mode}")
print(f"Standard Deviation CO: {std_dev}")
plt.figure(figsize=(10, 6))
sns.countplot(x=df['AQI_Bucket'], color='skyblue')
plt.title("Air Quality Index Count Plot")
plt.xlabel("Air Quality Index ")
plt.ylabel("Count")
plt.show()
```

```python
parameter = 'AQI'
parameter_data = df[parameter]

mean = parameter_data.mean()
median = parameter_data.median()
mode = parameter_data.mode()[0]
std_dev = parameter_data.std()

print(f"Mean {parameter}: {mean}")
print(f"Median {parameter}: {median}")
print(f"Mode {parameter}: {mode}")
print(f"Standard Deviation {parameter}: {std_dev}")

plt.figure(figsize=(10, 6))
plt.hist(parameter_data, bins=20, color='skyblue', edgecolor='black')
plt.title(f"{parameter} Distribution")
plt.xlabel('AQI (Satisfactory - severe)')
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='PM2.5', y='PM10', hue='AQI_Bucket', palette='viridis', alpha=0.7)
plt.title("Scatter Plot: PM2.5 vs. PM10 by AQI Bucket")
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.legend(title='AQI_Bucket')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
aqi_bucket_by_city = df.groupby(['City', 'AQI_Bucket'])['AQI_Bucket'].count().unstack(fill_value=0)

ax = aqi_bucket_by_city.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title("Distribution of AQI Buckets by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title='AQI Bucket')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
continuous_columns = [
    'Benzene',
    'Toluene',
    'Xylene'
]
corr_matrix = df[continuous_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap for Benzene, Toluene, and Xylene")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.lineplot(x='Date', y='PM2.5', data=df)
plt.title("PM2.5 over Time")
plt.subplot(2, 2, 2)
sns.lineplot(x='Date', y='CO', data=df)
plt.title("CO over Time")
plt.subplot(2, 2, 3)
sns.lineplot(x='Date', y='AQI', data=df)
plt.title("AQI over Time")
plt.subplot(2, 2, 4)
sns.boxplot(x='AQI_Bucket', y='AQI', data=df, palette='viridis')
plt.title("AQI Distribution by Bucket")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Separate DataFrame based on AQI_Bucket
category1_df = df[df['AQI_Bucket'] == 'Category 1 (Good)']
category2_df = df[df['AQI_Bucket'] == 'Category 2 (Satisfactory)']
category3_df = df[df['AQI_Bucket'] == 'Category 3 (Moderate)']
category4_df = df[df['AQI_Bucket'] == 'Category 4 (Poor)']
category5_df = df[df['AQI_Bucket'] == 'Category 5 (Very Poor)']

# Display the first few rows of each AQI bucket category
print("Category 1 (Good):")
print(category1_df.head())

print("\nCategory 2 (Satisfactory):")
print(category2_df.head())

print("\nCategory 3 (Moderate):")
print(category3_df.head())

print("\nCategory 4 (Poor):")
print(category4_df.head())

print("\nCategory 5 (Very Poor):")
print(category5_df.head())
```

```python
import matplotlib.pyplot as plt
aqi_bucket_distribution = df['AQI_Bucket'].value_counts()
explode = [0.2 if bucket == 'Severe' else 0 for bucket in aqi_bucket_distribution.index]
plt.figure(figsize=(8, 8))
plt.pie(aqi_bucket_distribution, labels=aqi_bucket_distribution.index, autopct='%1.1f%%',
        startangle=140, explode=explode, shadow=True)
plt.title("Distribution of AQI Buckets (3D)")
plt.show()
```

```python
# Separate DataFrame based on NO2 levels
low_no2_df = df[df['NO2'] < 20]  # You can adjust the threshold as needed
moderate_no2_df = df[(df['NO2'] >= 20) & (df['NO2'] < 40)]
high_no2_df = df[(df['NO2'] >= 40) & (df['NO2'] < 60)]
very_high_no2_df = df[df['NO2'] >= 60]

# Display the first few rows of each NO2 level category
print("Low NO2 Levels:")
print(low_no2_df.head())

print("\nModerate NO2 Levels:")
print(moderate_no2_df.head())

print("\nHigh NO2 Levels:")
print(high_no2_df.head())

print("\nVery High NO2 Levels:")
print(very_high_no2_df.head())
```
